## Optimización Bayesiana de los hiperparámetros

```python
# Uniforme y quniforme (tiene que ver con los cuartiles)
# Optimización bayesiana no toma todos los valores como si hace un gridsearch

# Definir rango de los hiperparámetros
space = {
    # Variar desde 3 hasta 18 con saltos de uno
    # se usa quniform por la distribución que no es normal
    "max_depth": hp.quniform('max_depth', 3, 18, 1),
    "gamma": hp.uniform("gamma", 1,9),
    "reg_alpha": hp.quniform('reg_alpha', 40, 180, 1), #Regularización
    "reg_lambda": hp.uniform('reg_lambda', 0, 1), #Regularización
    "colsample_bytree": hp.uniform('colsample_bytree', 0.5, 1),
    "min_child_weight": hp.quniform('min_child_weight', 0, 10, 1),
    # se puede usar constante
    "n_estimators": 500,
    "seed": 0
}
space
```

# Modificar estas funcioncitas para ponerle hiperparámetros
hp.choice(label, options)-> lista
hp.randint(label, upper)-> Valor random entre 0 y upper
hp.uniform(label, low, high)-> valor uniforme entre low y high
hp.quniform(label, low, high, q)-> round(uniform_(low/high)/q)*q
hp.normal(label, mean, std)-> Con distribución normal


```python
# Definir función objetivo
def objective(space):
    # Clasificador del xgboost # Clasificación
    clf = xgb.XGBClassifier(
        max_depth = int(space["max_depth"]),
        gamma = space["gamma"],
        reg_alpha = int(space["reg_alpha"]),
        reg_lambda = space["reg_lambda"],
        colsample_bytree = int(space["colsample_bytree"]),
        min_child_weight = int(space["min_child_weight"]),
        n_stimators = space["n_estimators"]
    ) 

    # Accuracy no sirve para sistemas desbalanceados PENDIENTE BALANCEAR O USAR ROC
    evaluation = [( X_train_prep, y_train), ( X_test_prep, y_test)]

    clf.fit(X_train_prep, y_train,
        eval_set=evaluation, eval_metric="auc",  # AUC es para 
        early_stopping_rounds=10,verbose=False)
    
    pred = clf.predict(X_test_prep)
    accuracy = accuracy_score(y_test, pred>0.5)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
# Algoritmo de optimización
trials = Trials()

best_hyperparams = fmin(
    fn=objective, # Función objetivo
    space=space, # Espacio donde se definen los hiperparámetros
    algo=tpe.suggest,
    max_evals=5, #Iteraciones
    trials=trials
)

# Investigar Euristico vs determinista
```

```python
# Cuales fueron los mejores hiperparámetros?
best_hyperparams
```

```python
# Final model 
xgb_model = XGBClassifier(
    max_depth = int(best_hyperparams["max_depth"]),
    gamma = best_hyperparams["gamma"],
    reg_alpha = int(best_hyperparams["reg_alpha"]),
    # reg_lambda = best_hyperparams["reg_lambda"],
    colsample_bytree = int(best_hyperparams["colsample_bytree"]),
    min_child_weight = int(best_hyperparams["min_child_weight"]),
    n_stimators = 500,
    scale_pos_weight = 1, # Si está desbalanceado aqui se corrige
    seed = 27
)
```
